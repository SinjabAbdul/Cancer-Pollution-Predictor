# Setup

```python
## YOUR CODE HERE
import pandas as pd
import numpy as np
```

```python
aqi_df = pd.read_csv('Data/API_M_2014_2018.csv')
cancer_df = pd.read_csv('Data/Mergered_CA_WA_OR.csv')
```

```python
#State data annually

# California
CA2014_aqi_df = aqi_df[52:104]
CA2015_aqi_df = aqi_df[1108:1159]
CA2016_aqi_df = aqi_df[2168:2220]
CA2017_aqi_df = aqi_df[3223:3274]
CA2018_aqi_df = aqi_df[4287:4338]
 
# Oregon
OR2014_aqi_df = aqi_df[737:758]
OR2015_aqi_df = aqi_df[1803:1824]
OR2016_aqi_df = aqi_df[2854:2874]
OR2017_aqi_df = aqi_df[3913:3934]
OR2018_aqi_df = aqi_df[4971:4992] 

# Washington
WA2014_aqi_df = aqi_df[966:993]
WA2015_aqi_df = aqi_df[2027:2054]
WA2016_aqi_df = aqi_df[3084:3111]
WA2017_aqi_df = aqi_df[4145:4173]
WA2018_aqi_df = aqi_df[5201:5229] 

```

# Merging the annual data for each state

California

```python
#merge to make CA df of 2014 - 2015 data
CA1415_aqi_df = CA2014_aqi_df.append(CA2015_aqi_df, ignore_index=True)

#merge to make CA df of 2014 - 2016 data
CA141516_aqi_df = CA1415_aqi_df.append(CA2016_aqi_df, ignore_index=True)

#merge to make CA df of 2014 - 2017 data
CA14151617_aqi_df = CA141516_aqi_df.append(CA2017_aqi_df, ignore_index=True)

#merge to make CA df of 2014 - 2018 data
CA1415161718_aqi_df = CA14151617_aqi_df.append(CA2018_aqi_df, ignore_index=True)
```

Oregon

```python
#merge to make OR df of 2014 - 2015 data
OR1415_aqi_df = OR2014_aqi_df.append(OR2015_aqi_df, ignore_index=True)

#merge to make OR df of 2014 - 2016 data
OR141516_aqi_df = OR1415_aqi_df.append(OR2016_aqi_df, ignore_index=True)

#merge to make OR df of 2014 - 2017 data
OR14151617_aqi_df = OR141516_aqi_df.append(OR2017_aqi_df, ignore_index=True)

#merge to make OR df of 2014 - 2018 data
OR1415161718_aqi_df = OR14151617_aqi_df.append(OR2018_aqi_df, ignore_index=True)
```

Washington

```python
#merge to make WA df of 2014 - 2015 data
WA1415_aqi_df = WA2014_aqi_df.append(WA2015_aqi_df, ignore_index=True)

#merge to make WA df of 2014 - 2016 data
WA141516_aqi_df = WA1415_aqi_df.append(WA2016_aqi_df, ignore_index=True)

#merge to make WA df of 2014 - 2017 data
WA14151617_aqi_df = WA141516_aqi_df.append(WA2017_aqi_df, ignore_index=True)

#merge to make WA df of 2014 - 2018 data
WA1415161718_aqi_df = WA14151617_aqi_df.append(WA2018_aqi_df, ignore_index=True)
```

```python
#merging CA 2014-18 with OR 2014-18
CA_OR_aqi_df = CA1415161718_aqi_df.append(OR1415161718_aqi_df, ignore_index=True)

#merging CA 2014-18 & OR 2014-18 with WA 2014-18 to create all of westcoast
westcoast_aqi_df = CA_OR_aqi_df.append(WA1415161718_aqi_df, ignore_index=True)


```

```python
westcoast_aqi_df.to_csv('westcoast_aqi_df.csv', index=False)
westcoast_aqi_df
```
